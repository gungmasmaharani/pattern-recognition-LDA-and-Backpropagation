```python
import numpy as np
import cv2 as lo
import matplotlib.pyplot as plt
import os
```

```python
def cari(img):
    p,l = img.shape
    x = bool(0)
    for i in range(100,p):
        for j in range(100,l):
            if(img[i,j]<20):
                m,n=i,j
                x = bool(1)
                break
        if (x):
            break
    return m,n
```

```python
def normal(img):
    p,l,t = img.shape
    img[:10,:]=255
    img[:,:7]=255
    img[:,(l-8):]=255
    img[(p-10):,:]=255
    return img
```

```python
def crop (img,i):
    x = img #[727:3850,368:(l-291)]
    #plt.imshow(x)
    i=str(i)
    c,v,n = x.shape
    ha = x[:int(c/3),:int(v/6)] 
    na = x[:int(c/3),int(v/6):int((2*v)/6)]
    ca = x[:int(c/3),int((2*v)/6):int((3*v)/6)]
    ra = x[:int(c/3),int((3*v)/6):int((4*v)/6)]
    ka = x[:int(c/3),int((4*v)/6):int((5*v)/6)]
    da = x[:int(c/3),int((5*v)/6):int(v-1)]

    ta = x[int(c/3):int(2*c/3),:int(v/6)] 
    sa = x[int(c/3):int(2*c/3),int(v/6):int((2*v)/6)]
    wa = x[int(c/3):int(2*c/3),int((2*v)/6):int((3*v)/6)]
    la = x[int(c/3):int(2*c/3),int((3*v)/6):int((4*v)/6)]
    ma = x[int(c/3):int(2*c/3),int((4*v)/6):int((5*v)/6)]
    ga = x[int(c/3):int(2*c/3),int((5*v)/6):int(v-1)]

    ba = x[int(2*c/3):int(c-1),:int(v/6)] 
    nga = x[int(2*c/3):int(c-1),int(v/6):int((2*v)/6)]
    pa = x[int(2*c/3):int(c-1),int((2*v)/6):int((3*v)/6)]
    ja = x[int(2*c/3):int(c-1),int((3*v)/6):int((4*v)/6)]
    ya = x[int(2*c/3):int(c-1),int((4*v)/6):int((5*v)/6)]
    nya = x[int(2*c/3):int(c-1),int((5*v)/6):int(v-1)]
    
    lo.imwrite('ha('+ i +').png', normal(ha))
    lo.imwrite('na('+ i +').png' , normal(na))
    lo.imwrite('ca('+ i +').png', normal(ca))
    lo.imwrite('ra('+ i +').png', normal(ra))
    lo.imwrite('ka('+ i +').png', normal(ka))
    lo.imwrite('da('+ i +').png', normal(da))
    lo.imwrite('ta('+ i +').png', normal(ta))
    lo.imwrite('sa('+ i +').png', normal(sa))
    lo.imwrite('wa('+ i +').png', normal(wa))
    lo.imwrite('la('+ i +').png', normal(la))
    lo.imwrite('ma('+ i +').png', normal(ma))
    lo.imwrite('ga('+ i +').png', normal(ga))
    lo.imwrite('ba('+ i +').png', normal(ba))
    lo.imwrite('nga('+ i +').png', normal(nga))
    lo.imwrite('pa('+ i +').png', normal(pa))
    lo.imwrite('ja('+ i +').png', normal(ja))
    lo.imwrite('ya('+ i +').png', normal(ya))
    lo.imwrite('nya('+ i +').png', normal(nya))
```

```python
path = "G:\\DATASET"
i=1
for r,d,f in os.walk(path):
    for file in f:
        if ".JPG" in file:
            img= lo.imread(os.path.join(path,file))
            crop(img,i)
            i+=1
        print(file)
```
