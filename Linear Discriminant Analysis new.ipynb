# LDA - Linear Discriminant Analysis

## Step wise LDA

### Step 1 - Computing d-dimensional mean vectors

```python
def kelas (file):
    if 'ha' in file: dot=1
    if 'na' in file: dot=2
    if 'ca' in file: dot=3
    if 'ra' in file: dot=4
    if 'ka' in file: dot=5
    if 'da' in file: dot=6
    if 'ta' in file: dot=7
    if 'sa' in file: dot=8
    if 'wa' in file: dot=9
    if 'la' in file: dot=10
    if 'ma' in file: dot=11
    if 'ga' in file: dot=12
    if 'ba' in file: dot=13
    if 'nga' in file: dot=14
    if 'pa' in file: dot=15
    if 'ja' in file: dot=16
    if 'ya' in file: dot=17
    if 'nya' in file: dot=18
    return dot
def norma (img):#normalisasi min max
    x,y = img.shape
    balik = np.zeros(img.shape)
    maks  = max(img.flatten())
    mins = min(img.flatten())
    for i in range(x):
        for j in range (y):
            balik[i,j]=(img[i,j]-mins)/(maks-mins)
    return balik.flatten()
```

```python
import os
import numpy as np
import cv2
path = "G:\\datatestLDA"
i=0
ID = []
dt=np.zeros((900,32*32))
for r,d,f in os.walk(path):
    for file in f:
        img= cv2.imread(os.path.join(path,file),0)#grayscale
        x = cv2.resize(img,(32,32))#resize
        dt[i,:]= norma(x)
        print(norma(x))
        ID.append(kelas (file))
        i+=1
    print(dt.shape)
```

```python
dt = np.array(dt)
ID = np.array(ID)
print(dt.shape, ID.shape)

print('data: ' + repr(dt),"\n")
print("ID :",ID,"\n")
```

```python
import sys
np.printoptions(threshold=sys.maxsize)

mean_vectors = []
for cl in range(1,19):
    mean_vectors.append(np.mean(dt[ID==cl], axis=0))
    print('Mean Vector class %s: %s\n' %(cl, mean_vectors[cl-1]))
```

### Step 2 - Computing the scatter matrices

#### 2.1 Within class scatter matrix

```python
S_W = np.zeros((1024,1024))
for cl,mv in zip(range(1,19), mean_vectors):
    class_sc_mat = np.zeros((1024,1024))                  # scatter matrix for every class
    for row in dt[ID == cl]:
        row, mv = row.reshape(1024,1), mv.reshape(1024,1) # make column vectors
        class_sc_mat += (row-mv).dot((row-mv).T)
    S_W += class_sc_mat                             # sum class scatter matrices
print('within-class Scatter Matrix: \n', S_W)
```

#### 2.2 Between class scatter matrix

```python
overall_mean = np.mean(dt, axis=0)
overall_mean.shape
S_W.shape
np.linalg.det(S_W)
```

```python
overall_mean = np.mean(dt, axis=0)

S_B = np.zeros((1024,1024))
for i,mean_vec in enumerate(mean_vectors):  #enumerate just adds a number from 0 to every value in object
    n = dt[y==i+1,:].shape[0]
    mean_vec = mean_vec.reshape(1024,1) # make column vector
    overall_mean = overall_mean.reshape(1024,1) # make column vector
    S_B += n * (mean_vec - overall_mean).dot((mean_vec - overall_mean).T)

print('between-class Scatter Matrix:\n', S_B)
```

### Step 3 - Solving the generalized eigenvalue problem for the matrix S−1WSB

```python
print(np.linalg.pinv(S_W))
eig_vals, eig_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B)) #makes eigenvector from matrix

for i in range(len(eig_vals)):
    eigvec_sc = eig_vecs[:,i].reshape(1024,1)   
    print('\nEigenvector {}: \n{}'.format(i+1, eigvec_sc.real))
    print('Eigenvalue {:}: {:.2e}'.format(i+1, eig_vals[i].real))
```

```python
#Checking the eigenvector-eigenvalue calculation
for i in range(len(eig_vals)):
    eigv = eig_vecs[:,i].reshape(4,1)
    np.testing.assert_array_almost_equal(np.linalg.inv(S_W).dot(S_B).dot(eigv),
                                         eig_vals[i] * eigv,
                                         decimal=6, err_msg='', verbose=True)
print('ok')
```

#### 4.1 Ordering eigenvectors by decreasing order of eigenvalue

```python
# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
```

```python
# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs = sorted(eig_pairs, key=lambda k: k[0], reverse=True)
```

```python
# Visually confirm that the list is correctly sorted by decreasing eigenvalues

print('Eigenvalues in decreasing order:\n')
for i in eig_pairs:
    print(i[0])
```

```python
#Representing values in %
print('Variance explained:\n')
eigv_sum = sum(eig_vals)
for i,j in enumerate(eig_pairs):
    print('eigenvalue {0:}: {1:.2%}'.format(i+1, (j[0]/eigv_sum).real))
```

#### 4.2 Choosing k eigenvectors with the largest eigenvalues

Here only 2 are adding information, hence selecting only first 2. We can also exclude the second one as the value add is too low.

```python
W = np.hstack((eig_pairs[0][1].reshape(4,1), eig_pairs[1][1].reshape(4,1))) #cbind 2 arrays
print('Matrix W:\n', W.real)
```
